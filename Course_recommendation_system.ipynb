```python
data = df[['Course Name','Difficulty Level','Course Description','Skills','University']]
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('punkt_tab')
nltk.download('punkt')
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def clean_text_nltk(text):
    # Tokenize the text
    word_tokens = word_tokenize(text)
    # Filter out punctuation and stop words
    filtered_text = [word for word in word_tokens if word.isalnum() and word not in stop_words]
    # Join words back to string
    cleaned_text = " ".join(filtered_text)
    return cleaned_text

# Apply the cleaning function to your DataFrame columns
data['Course Description'] = data['Course Description'].apply(clean_text_nltk)
data['Skills'] = data['Skills'].apply(clean_text_nltk)
```

```python
data['Course Name'] = data['Course Name'].apply(lambda x: x.replace(' ', ',').replace(',,', ',').replace(':', ''))

data['Course Description'] = data['Course Description'].apply(lambda x: x.replace(' ', ',')
                                                              .replace(',,', ',')
                                                              .replace('_', '')
                                                              .replace(':', '')
                                                              .replace('(', '')
                                                              .replace(')', ''))

data['Skills'] = data['Skills'].apply(lambda x: x.replace('(', '').replace(')', ''))
```

```python
data.head()
```

```python
data['Tags'] = data['Course Name'] + data['Difficulty Level'] + data['Course Description'] + data['Skills']+ data["University"]
```

```python
new_df = data[['Course Name','Tags']]
new_df.head()
```

```python
new_df['Tags'] = data['Tags'].str.replace(',',' ')
new_df['Course Name'] = data['Course Name'].str.replace(',',' ')
```

```python
new_df.rename(columns = {'Course Name':'course_name'}, inplace = True)
```

```python
new_df['course_name'] = new_df['course_name'].apply(lambda x:x.lower())
new_df['Tags'] = new_df['Tags'].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import word_tokenize
cv = CountVectorizer(max_features=5000, stop_words='english', ngram_range=(1, 2),
                     min_df=0.01, max_df=0.7, tokenizer=word_tokenize, binary=False)
```

```python
vectors = cv.fit_transform(new_df['Tags']).toarray()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
# Stemming = reducing a word to its root form.
def stem(text):
    # Ensure text is a string and not missing
    if not isinstance(text, str):
        return ""

    y = [ps.stem(i) for i in text.split()]  # List comprehension is often slightly faster
    return " ".join(y)

new_df['Tags'] = new_df['Tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
def recommend(course):
    # Normalize the input course name to lowercase
    course = course.strip().lower()
    # Find matching course index
    course_indices = new_df[new_df['course_name'].str.lower() == course].index
    if course_indices.size == 0:
        print(f"Course '{course}' not found. Please check the course name.")
        return

    # Get the index of the found course
    course_index = course_indices[0]
    # Fetch similarity scores
    distances = similarity[course_index]
    # Sort by similarity score in descending order (excluding the input course itself)
    course_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:7]

    # Display recommendations
    print(f"\nRecommended courses similar to '{course}':\n")
    for i, score in course_list:
        print(f"- {new_df.iloc[i]['course_name']}  (Similarity: {score:.2f})")
```

```python
recommend('Finance for Managers')
```

**Use Semantic Embeddings**

```python

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Load dataset
df1 = pd.read_csv("Coursera.csv")

# Keep only useful columns
data = df1[['Course Name', 'Course Description', 'Skills', 'University', 'Difficulty Level']]
data.fillna("", inplace=True)

# Combine everything into one searchable text
data["full_text"] = (
    data["Course Name"] + " " +
    data["Course Description"] + " " +
    data["Skills"] + " " +
    data["University"] + " " +
    data["Difficulty Level"]
)

# Use a pre-trained semantic embedding model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Create embeddings for all courses
embeddings = model.encode(data["full_text"], convert_to_tensor=False)

# Save everything
joblib.dump(data, "courses_df1.pkl")
joblib.dump(model, "embedding_model1.pkl")
joblib.dump(embeddings, "embeddings1.pkl")
```
